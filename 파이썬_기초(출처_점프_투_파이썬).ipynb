##Q2. while문을 사용해 1부터 1000까지의 자연수 중 3의 배수의 합을 구하세요

````python
# 답을 작성해주세요
sum = 0
i = 1

while i <= 1000:
  if i % 3 == 0:
    sum += i
    i += 1
  else:
    sum += 0
    i += 1

print(sum)

````

##Q1. 주어진 자연수가 홀수인지 짝수인지 판별하는 함수를 작성하세요

````python
# 답을 작성해주세요
def is_odd(a):
  if a % 2 == 0:
    return "짝수입니다"
  else:
    return "홀수입니다"

result = is_odd(2)
print(result)
````

##Q2. 입력으로 들어오는 모든 수의 평균을 계산해주는 함수를 작성하고 실행하세요
(입력으로 들어오는 수의 개수는 랜덤입니다.)

````python
# 답을 작성해주세요
def average_rand(n):
  sum = 0

  for i in n:
    sum += i
  
  length = len(n)
  avg = sum / length

  return avg

result = average_rand([1,2,3,4,5])
print (result)

````
